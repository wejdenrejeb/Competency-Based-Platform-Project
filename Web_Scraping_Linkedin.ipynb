## Create a function to extract Job Description from a linkedin profile in the Experience Section and See if the Owner of the profile studied at esprit or not

```python
def func1(url):
    from selenium import webdriver
    from bs4 import BeautifulSoup
    from selenium.webdriver.common.by import By
    import time
    import sys
    import pandas as pd
    import re

    ext1='/details/education'
    driver.get(url+ext1)
# ken l wifi mteek khayeb l 5 kaber fiha 
    time.sleep(8)
#implecitly
    html_source = driver.page_source
    
    fac="-"
    if "Ecole Supérieure Privée d'Ingénierie et de Technologies - ESPRIT" in html_source:
        fac="esprit"
        print("found")
    print(fac)
    
    

    extension = '/details/experience'
    driver.get(url+extension) 
    html_source = driver.page_source
    
    experience = None
    while experience is None:
        html_source = driver.page_source
        soup = BeautifulSoup(html_source, 'html.parser') 
        experience = soup.find('div', class_='scaffold-finite-scroll__content')



    list_csv=['poste','company_name','date','description','faculte']
    df = pd.DataFrame(columns=list_csv)

    exp=experience.find_all('li',{"pvs-list__paged-list-item artdeco-list__item pvs-list__item--line-separated"})
    for i in range(0,len(exp)):
        check=exp[i].find_all('li',{"pvs-list__paged-list-item"})
        if len(check) == 0:
            liste=[]
            poste=exp[i].find("span",{"mr1 t-bold"})
            poste=poste.find("span",{"aria-hidden":"true"}).get_text().strip().encode('UTF-16').decode('UTF-16')

            company_name=exp[i].find("span",{"t-14 t-normal"})
            company_name=company_name.find("span",{"aria-hidden":"true"}).get_text().strip().encode('UTF-16').decode('UTF-16')


            date=exp[i].find("span",{"t-14 t-normal t-black--light"})
            date=date.find("span",{"aria-hidden":"true"}).get_text().strip().encode('UTF-16').decode('UTF-16')



            description=exp[i].find("div",{"pvs-list__outer-container"})


            if description == None:
                description=None
            else :
                description=description.find("span",{"aria-hidden":"true"}).get_text().strip()
                description=re.sub(r'[^\w\s]', '', description)
                description= description.encode('UTF-16').decode('UTF-16')

                liste=[poste,company_name,date,description,fac]
                print(liste)
                for s in range(5 - len(liste)):       
                        liste=liste+[None]
                df = df.append(pd.Series(liste, index=df.columns), ignore_index=True)
            
        else :
            company_name=exp[i].find("span",{"mr1 hoverable-link-text t-bold"}).get_text().strip().encode('UTF-16').decode('UTF-16')
            for j in range(0,len(check)):

                  liste=[]
                  poste=check[j].find("span",{"mr1 hoverable-link-text t-bold"})

                  poste=poste.find("span",{"aria-hidden":"true"}).get_text().strip().encode('UTF-16').decode('UTF-16')






                  date=check[j].find("span",{"t-14 t-normal t-black--light"})
                  date=date.find("span",{"aria-hidden":"true"}).get_text().strip().encode('UTF-16').decode('UTF-16')



                  description=check[j].find("div",{"pvs-list__outer-container"})
                  if description == None:
                        description=None

                  else :
                    description=description.find("span",{"aria-hidden":"true"}).get_text().strip()
                    print(description)
#houni kol mara zidou l hajet li yodhhroulkom fihom erreuret bech maadech ikalkoukom
                    description=re.sub(r'[^\w\s]', '', description)
                    description=description.encode('UTF-16').decode('UTF-16')


                    liste=[poste,company_name,date,description,fac]
                    print(liste)
               
                    for s in range(5 - len(liste)):       
                        liste=liste+[None]
                    df = df.append(pd.Series(liste, index=df.columns), ignore_index=True)

    with open('data_pi.csv', 'a') as f:
            df.to_csv(f, index=False,line_terminator='\n',encoding="UTF-16",header=None)
    return df
```
